```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

data = pd.read_csv('car_stopping_distance.csv')

x = data[['Speed (mph)']]
y = data['Distance (ft)']

def fitting(a, b1):
    y_pred = a + b1*x
    
    plt.plot(x, y, 'o')
    plt.plot(x, y_pred)
    plt.xlabel('Speed (mph)')
    plt.ylabel('Distance (ft)')
    plt.title(f'$y = {a} + {b1}x$')
    plt.show()

interact(fitting, a=(-11, 0), b1=(1, 5, .1))
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

xp = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x)

def plotting(a, b1, b2):
    y_pred = a + b1*xp[:, 0] + b2*xp[:, 1] 
    
    plt.plot(x, y, 'o')
    plt.plot(x, y_pred)
    plt.xlabel('Speed (mph)')
    plt.ylabel('Distance (ft)')
    plt.title(f'$y = {a} + {b1}x + {b2}x^2$')
    plt.show()

interact(plotting, a=(1, 10), b1=(0, 1, .1), b2=(0, .1, .01))
plt.show()
```
